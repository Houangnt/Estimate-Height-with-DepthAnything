```python
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt

def calculate_segment_area(image_path, model_path, conf=0.45):
    model = YOLO(model_path)

    img = cv2.imread(image_path)
    if img is None:
        print("Error: Could not load image")
        return 0, None, None

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    original_height, original_width = img.shape[:2]

    results = model.predict(image_path, conf=conf)
    result = results[0]

    if result.masks is None:
        print("No segmentation mask found")
        return 0, None, None

    # Get the class ID
    if len(result.boxes) > 0:
        class_id = int(result.boxes[0].cls.item() + 1)
        stage = f"Giai Ä‘oáº¡n {class_id}"
    else:
        stage = "KhÃ´ng xÃ¡c Ä‘á»‹nh Ä‘Æ°á»£c giai Ä‘oáº¡n"

    # Get the mask for the first detection
    mask = result.masks.data[0].cpu().numpy()

    # Convert mask to binary image (0 and 255)
    mask = (mask * 255).astype(np.uint8)

    # Resize mask to match original image dimensions
    mask = cv2.resize(mask, (original_width, original_height), interpolation=cv2.INTER_NEAREST)

    # Count non-zero pixels
    pixel_count = cv2.countNonZero(mask)

    colored_mask = np.zeros_like(img)
    colored_mask[:, :, 1] = mask

    # Overlay mask on original image
    alpha = 0.5
    overlay = cv2.addWeighted(img, 1, colored_mask, alpha, 0)

    return img, overlay, pixel_count, stage

def display_results(image_path, model_path, save_path=None):
    # Calculate area and get visualizations
    original, overlay, area_pixels, growth_stage = calculate_segment_area(image_path, model_path)

    # Print results
    print(f"Diá»‡n tÃ­ch (pixels): {area_pixels}")
    print(f"PhÃ¢n loáº¡i: {growth_stage}")

    # Create a figure with original and overlay
    plt.figure(figsize=(12, 5))

    # Original Image
    plt.subplot(121)
    plt.title('áº¢nh gá»‘c')
    plt.imshow(original)
    plt.axis('off')

    # Overlay
    plt.subplot(122)
    plt.title(f'Káº¿t quáº£ phÃ¢n Ä‘oáº¡n\n{growth_stage}')
    plt.imshow(overlay)
    plt.axis('off')

    plt.tight_layout()

    # Save the figure if save_path is provided
    if save_path:
        plt.savefig(save_path, bbox_inches='tight', dpi=300)
        print(f"Results saved to {save_path}")

    plt.show()

# Example usage
image_path = "/content/test/RGB_199.png"
model_path = "/content/lettuce.pt"
save_path = "/content/results.png"

display_results(image_path, model_path, save_path)
```
